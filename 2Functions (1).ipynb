##2. Difference between Mutable and immutable data types?
- Mutable
   - It can be changed after creation.
   - Same objects can be updated.
   - Example- list, dic, set
   ```
   list = [1, 2, 3]print(list)    # [10, 2, 3]
   ```
-  Immutable
   -  Cannot be changed once created.
   -  New object is created if changed.
   - Example- int, float, string, tuple
   ```     
   string = "hello"
   #string[0] = "H"
   new_string = "H" + string[1:]  # create a new string
   print(new_string)  # "Hello"
   ```

##8. What is a hash table and how does it relate to dictionaries in Python?

##10. Why are strings immutable in Python?
   -  Strings are immutable → you can’t change characters once created. -  
  -  Reasons:

  1. Efficiency: Immutable objects are faster for reuse.
  2. Security: If strings could change, it might cause bugs in programs.
  3. Hashing: Strings are used as dictionary keys; immutability ensures they
     don’t change unexpectedly.

##11. Advantages of dictionaries over lists.?
   - Lists → data is accessed by **index number**.
   - Dictionaries  → data is accessed by **keys**, which are more meaningful.

- Advantages of dictionaries:
    - Easier to understand (keys describe the data).
    - Faster lookups (uses hash table).
    - Can handle large, structured data better.

## 14. How does the "in" keyword work differently for lists and dictionaries?
   - In lists → checks if a value exists.

    ```
     list = [1, 2, 3]
     print(2 in list)   # True
     ```

   - In dictionaries → checks if a key exists (not value).

     ```
     dict = {"name": "Tanuj", "age": 21}
     rint("name" in dict)   # True
     print("Tanuj" in dict)  # False (because it’s a value)
     ```

##16. What is a nested dictionary? Example use case.?
  -  A **nested dictionary** = dictionary inside another dictionary.
  - Example: storing student data.

```
students = {
    "101": {"name": "Tanuj", "age": 200}
    "102": {"name": "Ravi", "age": 200}
}
print(students["101"]["name"])  # Tanuj
```
##17. Time complexity of accessing elements in a dictionary.?
  - Accessing elements in a dictionary is on average O(1) → very fast.
   This is because dictionaries use a hash table.
   Worst case can be O(n) (rare, if many hash collisions happen).

##18. Situations where lists are preferred over dictionaries.?
   - When you only need an **ordered collection** of items.
    When **keys are not important**, just the values.
    Example:

  -  A shopping list → `["milk", "bread", "eggs"]`.
  -  A series of marks → `[78, 85, 92]`.

##19. Why are dictionaries considered unordered, and how does that affect   data retrieval?
   - In Python (before 3.7), dictionaries were **unordered** → items had no
     fixed position.
   -  From Python 3.7+, they maintain **insertion order**, but they are still
     called “unordered” because:

  - Data is stored using a **hash table**, not by index.
      Effect:
  -  You **cannot access by index** (`dict[0] ❌`).
  -  You must access by **key**.

29. Write a code to combine two dictionaries into one and print the result.?

```py
# create two dictionaries
dict1 = {"name": "Tanuj", "age": 101}
dict2 = {"city": "Paris", "country": "France"}

# create a new empty dictionary
combined = {}

# add all items of dict1
for key in dict1:
    combined[key] = dict1[key]

# add all items of dict2
for key in dict2:
    combined[key] = dict2[key]

# print the combined dictionary
print(combined)
```
Output:

```
{'name': 'Tanuj', 'age': 101, 'city': 'Paris', 'country': 'France'}
```

30. Convert a list of strings into a set.?

```python
# create a list of strings
fruits = ["apple", "banana", "mango", "apple", "banana"]

# create an empty set
fruit_set = set()

# add each element of the list into the set
for item in fruits:
    fruit_set.add(item)

# print the set
print(fruit_set)
```

Output:    # order may change, and duplicates are removed

```
{'banana', 'apple', 'mango'}
```
